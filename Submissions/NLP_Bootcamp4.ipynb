# **Imports**

```python
import os
import sys
import tensorflow as tf
from datasets import load_dataset
from transformers import AutoTokenizer, TFAutoModelForSeq2SeqLM, DataCollatorForSeq2Seq, Seq2SeqTrainingArguments, Seq2SeqTrainer, AdamWeightDecay
from transformers import Seq2SeqTrainingArguments, Seq2SeqTrainer
import gradio as gr
```

# **Loading Dataset and Pre-trained Transformer**

```python
dataset = load_dataset("opus100", "en-fr")
model_name = "Helsinki-NLP/opus-mt-en-fr"
```

# **Tokenizer**

```python
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
tokenizer("Hello World, I am Rakshit!")
```

```python
tokenizer(["Hello World, I am Rakshit!", "This is final of NLP Bootcamp."])
```

# **Pre-Processing**

```python
max_input_length = 128
max_target_length = 128

source_lang = "en"
target_lang = "fr"
```

```python
def preprocess_function(examples):
    inputs = [ex[source_lang] for ex in examples["translation"]]
    targets = [ex[target_lang] for ex in examples["translation"]]
    model_inputs = tokenizer(inputs, max_length=max_input_length, truncation=True)

    # Setup the tokenizer for targets
    with tokenizer.as_target_tokenizer():
        labels = tokenizer(targets, max_length=max_target_length, truncation=True)

    model_inputs["labels"] = labels["input_ids"]
    return model_inputs
```

```python
preprocess_function(dataset["train"][:2])
```

```python
tokenized_datasets = dataset.map(preprocess_function, batched=True)
```

```python
model = TFAutoModelForSeq2SeqLM.from_pretrained(model_name)
```

# **Dataset Preparation**

```python
batch_size = 16
learning_rate = 2e-5
weight_decay = 0.01
num_train_epochs = 1
```

```python
data_collator = DataCollatorForSeq2Seq(tokenizer, model=model, return_tensors="tf")
```

```python
generation_data_collator = DataCollatorForSeq2Seq(tokenizer, model=model, return_tensors="tf", pad_to_multiple_of=128)
```

```python
train_dataset = model.prepare_tf_dataset(
    tokenized_datasets["test"],
    batch_size=batch_size,
    shuffle=True,
    collate_fn=data_collator,
)
```

```python
validation_dataset = model.prepare_tf_dataset(
    tokenized_datasets["validation"],
    batch_size=batch_size,
    shuffle=False,
    collate_fn=data_collator,
)
```

```python
generation_dataset = model.prepare_tf_dataset(
    tokenized_datasets["validation"],
    batch_size=8,
    shuffle=False,
    collate_fn=generation_data_collator,
)
```

# **Training**

```python
optimizer = AdamWeightDecay(learning_rate=learning_rate, weight_decay_rate=weight_decay)
model.compile(optimizer=optimizer)
```

```python
model.fit(train_dataset, validation_data=validation_dataset, epochs=1)
```
